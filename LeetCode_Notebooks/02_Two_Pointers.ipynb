```python
class Solution:
    # Problem 1: Valid Palindrome
    def isPalindrome(self, s: str) -> bool:
        s = ''.join(filter(str.isalnum, s)).lower()  # Remove non-alphanumeric and convert to lowercase
        left, right = 0, len(s) - 1
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True

    # Problem 2: 3Sum
    def threeSum(self, nums):
        nums.sort()
        result = []
        for i in range(len(nums) - 2):
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            left, right = i + 1, len(nums) - 1
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])
                    left += 1
                    right -= 1
                    while left < right and nums[left] == nums[left - 1]:
                        left += 1
                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1
                elif total < 0:
                    left += 1
                else:
                    right -= 1
        return result

    # Problem 3: Container With Most Water
    def maxArea(self, height):
        left, right = 0, len(height) - 1
        max_area = 0
        while left < right:
            width = right - left
            max_area = max(max_area, width * min(height[left], height[right]))
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area


# Create an instance of Solution
solution = Solution()

# Example for Valid Palindrome
print("Valid Palindrome:")
example1 = "A man, a plan, a canal: Panama"
print(f"Input: {example1}")
print(f"Output: {solution.isPalindrome(example1)}\n")

# Example for 3Sum
print("3Sum:")
example2 = [-1, 0, 1, 2, -1, -4]
print(f"Input: {example2}")
print(f"Output: {solution.threeSum(example2)}\n")

# Example for Container With Most Water
print("Container With Most Water:")
example3 = [1, 8, 6, 2, 5, 4, 8, 3, 7]
print(f"Input: {example3}")
print(f"Output: {solution.maxArea(example3)}")
```
