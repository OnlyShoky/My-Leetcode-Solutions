```python
class Solution:
    # Problem 1: Best Time to Buy and Sell Stock
    def maxProfit(self, prices) -> int:
        min_price = float('inf')
        max_profit = 0
        for price in prices:
            if price < min_price:
                min_price = price
            elif price - min_price > max_profit:
                max_profit = price - min_price
        return max_profit

    # Problem 2: Longest Substring Without Repeating Characters
    def lengthOfLongestSubstring(self, s: str) -> int:
        char_set = set()
        left = 0
        max_length = 0
        for right in range(len(s)):
            while s[right] in char_set:
                char_set.remove(s[left])
                left += 1
            char_set.add(s[right])
            max_length = max(max_length, right - left + 1)
        return max_length

    # Problem 3: Longest Repeating Character Replacement
    def characterReplacement(self, s: str, k: int) -> int:
        char_count = {}
        left = 0
        max_count = 0
        max_length = 0
        for right in range(len(s)):
            char_count[s[right]] = char_count.get(s[right], 0) + 1
            max_count = max(max_count, char_count[s[right]])
            while (right - left + 1) - max_count > k:
                char_count[s[left]] -= 1
                left += 1
            max_length = max(max_length, right - left + 1)
        return max_length
    
    def minWindow(self, s: str, t: str) -> str:
        result = []
        for i,c in enumerate(s) :
            combo = (c,i),
            print(combo)
            if c in t :
                # result.append((c,i))
                combo = (c,i),
                print(combo)
        
        return result


# Create an instance of Solution
solution = Solution()

# Example for Best Time to Buy and Sell Stock
print("Best Time to Buy and Sell Stock:")
example1 = [7, 1, 5, 3, 6, 4]
print(f"Input: {example1}")
print(f"Output: {solution.maxProfit(example1)}\n")

# Example for Longest Substring Without Repeating Characters
print("Longest Substring Without Repeating Characters:")
example2 = "abcabcbb"
print(f"Input: {example2}")
print(f"Output: {solution.lengthOfLongestSubstring(example2)}\n")

# Example for Longest Repeating Character Replacement
print("Longest Repeating Character Replacement:")
example3 = ("AABABBA", 1)
print(f"Input: {example3}")
print(f"Output: {solution.characterReplacement(*example3)}\n")

# Example for Minimum Window Substring
print("Minimum Window Substring:")
s = "OUZODYXAZV" 
t = "XYZ"
print(f"Input: {s} , { t}")
print(f"Output: {solution.minWindow(s,t)}\n")
```

```python
from collections import Counter
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        
        if len(s) < len(t) or not t or not s:
            return ""

        result = []
        finalResult = []
        track = Counter(t)
        progress = t[:]
        for i,c in enumerate(s) :

            if c in t :
                track[c] -=1
                result.append((c,i))
                progress =progress.replace(c,'',1)

            while result and track[result[0][0]] < 0 :
                track[result[0][0]] +=1
                result = result[1:]
                
            if progress == '' :
                if not finalResult :
                    finalResult = result.copy()
                else:
                    distResult = result[len(result)-1][1]   - result[0][1]     
                    distFinal = finalResult[len(finalResult)-1][1]   -  finalResult[0][1]    
                    if(distResult < distFinal):\
                        finalResult = result.copy()


        if not result or progress != '' :
            return ''
        

        
        if len(finalResult)  == 1:
            return s[finalResult[0][1]]
        else :
            l = finalResult[0][1]     
            r = finalResult[len(finalResult)-1][1]  
            return s[l:r+1]
              
            
        
        


# Create an instance of Solution
solution = Solution()


# Example for Minimum Window Substring
print("Minimum Window Substring:")
s = "cabwefgewcwaefgcf"
t = "cae"

s = "babb"
t = "baba"
print(f"Input: {s} , { t}")
print(f"Output: {solution.minWindow(s,t)}\n")
```
